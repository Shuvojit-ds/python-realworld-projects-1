```python
#Smart parking system
empty_space=0
parking_capacity=10

while empty_space<=parking_capacity:
  entry=input("Is there car is entering(Yes/No):")
  if entry=="Yes":
    empty_space=empty_space+1
    Available_space=parking_capacity-empty_space
    print("Available_space:",Available_space)
  elif Available_space==0:
     print("Sorry capacity is full")
     break
```
